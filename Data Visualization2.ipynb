```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.preprocessing import Normalizer
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
normal_df=pd.DataFrame({'x':np.random.randint(-100,100,1000),
                        'y':np.random.randint(-80,80,1000),
                        'z':np.random.randint(-150,150,1000)})
```

```python
normal_df
```

```python
import matplotlib.pyplot as plt
```

```python
ax=plt.axes(projection='3d')
ax.scatter3D(normal_df.x,normal_df.y,normal_df.z)
plt.figure()
```

```python
norm=Normalizer()
```

```python
norm_df=norm.fit_transform(normal_df)
```

```python
norm_df
```

```python
norm_df=pd.DataFrame(normal_df,columns=['x','y','z'])
```

```python
ax=plt.axes(projection='3d')
ax.scatter3D(norm_df.x,norm_df.y,norm_df.z)
plt.figure()
```

```python
data={'First Score':[100,90,np.nan,95],
      'Second Score':[98,90,59,np.nan],
      'Third Score':[80,99,88,95]}
```

```python
var=pd.DataFrame(data)
```

```python
var
```

```python
var.isnull()
var.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
imputer.fit(var)
imputer.transform(var)
```
